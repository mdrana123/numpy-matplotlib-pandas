### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

A figure in matplotliv is a window . It is highest level of container and can contain multiple plot.
We use method ( fig = plt.figure()   or fig = plt.subplots() ) to draw a figure.

A axes is plot elements where all the data is placed. In a figure can be multiple axes (subplots) and subplots can be separate plot. Examples: fig, ax = plt.subplots()
                             ax.plot([1, 2, 3], [4, 5, 6])

Axis is always refer to x or y of a object. Plots should be multiple axis objects.. Till examples:
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

Artist is an object that  can be drawn in a figure. Till examples:
line = ax.plot([1, 2, 3], [4, 5, 6])  # The line is an artist

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4, 4)
y = x**2

plt.plot(x,y)

plt.xlabel("x-label")
plt.ylabel("y-label")
plt.title("figure Y =x^2")

plt.show()
```
